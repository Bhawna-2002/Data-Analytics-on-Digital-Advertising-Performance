```python
import pandas as pd
```

```python
df=pd.read_excel("Data Analytics Assignment.xlsx",header=None)
```

```python
df.head(20)
```

```python
df.drop(0,axis=1,inplace=True)
```

```python
df.drop([0,1],axis=0,inplace=True)
```

```python
df.drop([5,6,14,15],axis=0,inplace=True)
```

```python

df.head(50)
```

```python
print(df.columns)
```

```python
# Step 1: Set correct header (row 1)
df.columns = df.iloc[1].values   # row index 1 has actual column names

# Step 2: Keep only data rows, reset index
df = df.iloc[2:].reset_index(drop=True)

# Step 3: Strip whitespace from column names
df.columns = df.columns.str.strip()

# Step 4: Convert numeric columns safely
numeric_cols = ['unique_idfas','unique_ips','unique_uas','total_requests',
                'requests_per_idfa','impressions','impressions_per_idfa',
                'idfa_ip_ratio','idfa_ua_ratio','IVT']

for col in numeric_cols:
    if col in df.columns:          # ensure column exists
        df[col] = pd.to_numeric(df[col], errors='coerce')  # invalid values -> NaN


# Step 5: Convert Date column
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.head()
```

```python
df[numeric_cols] = df[numeric_cols].fillna(0)
df.head()
```

```python
## computing the ivt flag
ivt_threshold = 0.01  # adjust based on your data
df['ivt_flag'] = df['IVT'] > ivt_threshold
```

```python
## time series plot
import matplotlib.pyplot as plt
import seaborn as sns

metrics = ['requests_per_idfa','idfa_ip_ratio','idfa_ua_ratio','impressions_per_idfa','IVT']

for metric in metrics:
    plt.figure(figsize=(12,5))
    sns.lineplot(data=df, x='Date', y=metric)
    plt.title(f'{metric} Over Time')
    plt.xticks(rotation=45)
    plt.show()
```

```python
## Boxplots to detect outliers
for metric in ['idfa_ua_ratio','requests_per_idfa','impressions_per_idfa']:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[metric])
    plt.title(f'Boxplot of {metric}')
    plt.show()
```

```python
## metrices related to IVT Correlation Analysis
corr = df[['idfa_ua_ratio','idfa_ip_ratio','requests_per_idfa','impressions_per_idfa','IVT']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
## summary of ivt events
ivt_summary = df[df['ivt_flag']][['Date','unique_idfas','unique_ips','unique_uas',
                                   'total_requests','requests_per_idfa',
                                   'idfa_ip_ratio','idfa_ua_ratio','IVT']].sort_values('Date')

ivt_summary.to_csv('ivt_summary_report.csv', index=False)
print("IVT summary report saved as 'ivt_summary_report.csv'")

# Optional: see top suspicious events
print(df.sort_values('idfa_ua_ratio', ascending=False).head(10))
```
